# Red wine quality prediction

Use XGBoost to demonstrate the calculation of permutation importances, partial dependency plots, ice, and SHAP.

```python
from sklearn.model_selection import train_test_split
from sklearn.inspection import permutation_importance # https://github.com/scikit-learn/scikit-learn/blob/7e1e6d09b/sklearn/inspection/_permutation_importance.py
import ice # https://github.com/AustinRochford/PyCEbox/blob/master/pycebox/ice.py
import shap

from xgboost import XGBRegressor

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
%matplotlib inline
```

## Load the data

```python
data = pd.read_csv('./data/winequality-red.csv')
```

```python
data.head()
```

## Pre-process the data

1. Check missing data - none
2. Drop the irrelevant columns - unecessary in this case
3. Ensure data types are correct
3. Encode the categorical columns
4. Normalize the continuous columns
5. Put back the categorical and continuous columns together
6. Split the data into a training set and a validation set

```python
# only continuous columns exist
X_all = data.drop('quality', axis = 1).values
X_all = (X_all - X_all.mean()) / X_all.std()
y_all = data['quality'].values

# train/test split
X_train, X_test, y_train, y_test = train_test_split(X_all, y_all, test_size = 0.2, random_state = 0)
```

## XGBoost training

```python
# n_estimators - number of trees in the forest
# max_depth - maximum depth of each regression tree
# eta - learning rate
# gamma - minimum loss reduction required to make a further partition on a leaf node of the tree
# subsample - fraction of samples to be used for fitting each tree
# colsample_bytree - fraction of features to be used for each tree
reg = XGBRegressor(n_estimators=500, max_depth=7, eta=0.1, gamma = 0, 
                   subsample=0.6, colsample_bytree=0.5, random_state = 333)
reg.fit(X_train, y_train)
print("MSE = {}".format(reg.score(X_test, y_test)))
y_val = reg.predict(X_test)

fig, ax = plt.subplots()
ax.plot(y_val, y_test, '.')
corr = np.corrcoef(y_val.reshape(-1), y_test)[0,1]
ax.text(0.6,0.1,f'Pearson corr = {corr:.6f}',transform = ax.transAxes)
```

## Permutation feature importance

```python
result = permutation_importance(reg, X_test, y_test, n_repeats = 10, random_state = 333)

fig, ax = plt.subplots()
ax.bar(range(X_train.shape[1]), result.importances_mean, yerr=result.importances_std)
ax.set_xticks(range(X_train.shape[1]))
ax.set_xticklabels(data.drop('quality', axis = 1).columns, rotation = 90)
ax.set_title("Feature importances using permutation")
ax.set_ylabel("Mean decrease in score")
```

## Partial Dependence Plot and Individual Conditional Expectations

```python
foi = 'alcohol'
ice_data = ice.ice(
    pd.DataFrame(X_train, columns = data.drop('quality', axis = 1).columns),
    foi, reg.predict, 50)
ice.ice_plot(ice_data, color_by = 'sulphates', cmap = 'viridis', lw = 0.2)
ax.set_title(f'ICE for feature {foi}')
```

```python
ice_data.head().T
```

```python
fig, axes = plt.subplots(1, 2, figsize = (12, 6))
for i, foi in enumerate(['alcohol', 'sulphates']):
    ice_data = ice.ice(
        pd.DataFrame(X_train, columns = data.drop('quality', axis = 1).columns),
        foi, reg.predict, 50)
    ax = axes[i]
    # viridis: yellow - high, purple - low
    ice.ice_plot(ice_data, frac_to_plot = 0.2, 
                 color_by = 'volatile acidity', cmap = 'viridis', ax = ax, lw = 0.2)
    ax.set_title(f'ICE for feature {foi}')
```

```python
fig, axes = plt.subplots(1, 2, figsize = (12, 6))
for i, foi in enumerate(['alcohol', 'sulphates']):
    ice_data = ice.ice(
        pd.DataFrame(X_train, columns = data.drop('quality', axis = 1).columns),
        foi, reg.predict, 50)
    ax = axes[i]
    # viridis: yellow - high, purple - low
    ice.ice_plot(ice_data, frac_to_plot = 0.2, plot_pdp = True, 
                 centered = True, centered_quantile = 0.,
                 color_by = 'volatile acidity', cmap = 'viridis',
                 ax = ax, lw = 0.2, pdp_kwargs = {'lw': 2, 'color': 'r'})
    ax.set_title(f'ICE for feature {foi}')
```

## SHAP

```python
explainer = shap.Explainer(reg)
shap_values = explainer(pd.DataFrame(X_train, columns = data.drop('quality', axis = 1).columns))
```

```python
# visualize the first prediction's explanation
shap.plots.waterfall(shap_values[0])
```

```python
# visualize the first prediction's explanation with a force plot
shap.initjs()
shap.plots.force(shap_values[0])
```

```python
# visualize all the training set predictions
shap.initjs()
shap.plots.force(shap_values)
```

```python
# create a dependence scatter plot to show the effect of a single feature across the whole dataset
shap.plots.scatter(shap_values[:, 'alcohol'], color=shap_values[:, 'sulphates'])
```

```python
# summarize the effects of all the features
shap.plots.beeswarm(shap_values)
```
